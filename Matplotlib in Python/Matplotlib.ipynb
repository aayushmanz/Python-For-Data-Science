```python
# from a pandas situation

batsman = pd.read_csv('/workspaces/Python-For-Data-Science/Matplotlib in Python/Dataset for matplotlib/sharma-kohli.csv')
batsman

plt.plot(batsman['index'], batsman['V Kohli'])
```

```python
# plotting mutiple plots
plt.plot(batsman['index'], batsman['V Kohli'])
plt.plot(batsman['index'], batsman['RG Sharma'])
```

```python
# label title
plt.plot(batsman['index'], batsman['V Kohli'])
plt.plot(batsman['index'], batsman['RG Sharma'])

plt.title("Rohit vs Virat career comparison")
plt.xlabel('Seasons')
plt.ylabel('Runs Scored each year')
```

```python
# colors(hex) and line(width and style) and marker(size)
plt.plot(batsman['index'], batsman['V Kohli'], color='red')
plt.plot(batsman['index'], batsman['RG Sharma'], color='blue')

plt.title("Rohit vs Virat career comparison")
plt.xlabel('Seasons')
plt.ylabel('Runs Scored each year')
```

```python
plt.plot(batsman['index'], batsman['V Kohli'], color='red', linestyle='solid', linewidth='2')
plt.plot(batsman['index'], batsman['RG Sharma'], color='blue', linestyle='dashed', linewidth='0.5')

plt.title("Rohit vs Virat career comparison")
plt.xlabel('Seasons')
plt.ylabel('Runs Scored each year')
```

```python
plt.plot(batsman['index'], batsman['V Kohli'], color='red', linestyle='dotted', marker='D', markersize='10')
plt.plot(batsman['index'], batsman['RG Sharma'], color='blue', linestyle='dashdot', marker='.')  
# +, > , < , D, o etc can use for marker.
# we can also controll the size of marker by markersize=

plt.title("Rohit vs Virat career comparison")
plt.xlabel('Seasons')
plt.ylabel('Runs Scored each year')
```

```python
# legend -> location

plt.plot(batsman['index'], batsman['V Kohli'], color='red', linestyle='solid', linewidth='2', label='Virat')
plt.plot(batsman['index'], batsman['RG Sharma'], color='blue', linestyle='dashed', linewidth='0.5', label='Rohit')
# plt.legend() is default
plt.legend(loc='upper right')
# legend is use for labelling the name of graph.

plt.title("Rohit vs Virat career comparison")
plt.xlabel('Seasons')
plt.ylabel('Runs Scored each year')
```

```python
# limiting axes 

price = [48000,54000, 57000, 49000,47000, 45000, 450000]
year = [2025, 2026, 2027, 2028, 2029, 2030, 2031]

plt.plot(year, price)
plt.ylim(0, 180000)
plt.xlim(2027, 2029)
```

```python
# grid :

plt.plot(batsman['index'], batsman['V Kohli'], color='red', linestyle='solid', linewidth='2', label='Virat')
plt.plot(batsman['index'], batsman['RG Sharma'], color='blue', linestyle='dashed', linewidth='0.5', label='Rohit')
plt.legend() # is default
# legend is use for labelling the name of graph.

plt.title("Rohit vs Virat career comparison")
plt.xlabel('Seasons')
plt.ylabel('Runs Scored each year')


plt.grid() # is for lineing in graph
```
