```python
# import titanic

df = pd.read_csv('/workspaces/Python-For-Data-Science/Database from Kaggle/titanic.csv')
df['Name']
```

```python
# common functions
# lower/upper/capitalize/title
df['Name'].str.lower()
df['Name'].str.upper()
df['Name'].str.capitalize()
df['Name'].str.title() # title case is for capital the first word of the sentence in starting.
# len
df["Name"][df["Name"].str.len() == 82].values[0]
```

```python
df['Name'].str.strip()
```

```python
# split -> get

df['lastname']= df["Name"].str.split(',').str.get(0)
df.head(1)
```

```python
df[['title', 'firstname']]= df['Name'].str.split(',').str.get(1).str.strip().str.split(' ',n=1,expand=True)
# n = 1 is use for controll spliting method.

df['title'].value_counts()
```

```python
# replace :


df['title'] = df['title'].str.replace({
    'Ms.':'Miss.',
    'Mlle.':'Miss.'

})

df['title'].value_counts()
```
