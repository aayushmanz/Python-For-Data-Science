# Timestamp Object :

* Time stamps reference particular moments in time (e.g, Oct 24th , 2026 at 6:00 pm)

* Creating Timestamp objects

```python
# Creating a timestamp
pd.Timestamp("16/02/2004")
```

```python
type(pd.Timestamp("16/02/2004"))
```

```python
# variations
pd.Timestamp('2004/02/16')
pd.Timestamp('2004, 2,16')
pd.Timestamp("16-02-2004")
```

```python
# only year
pd.Timestamp('2004')
```

```python
# using text
pd.Timestamp('1 Oct 2026')
```

```python
# providing time also
pd.Timestamp('5th january 2026 10:15am')
```
