**why separate objects to handle data and time when python already has datetime functionality?**

* syntax wise datetime is very convenient  
* But the performance takes a hit while working with huge data. List vs Numpy Array  
* The weakness of Python's datetime format inspired the NumPy team to add a set of native time series data type to NumPy.  
* The datetime64 dtype encodes dates as 64-bit integers, and thus allows arrays of dates to be represented very compactly.

```python
import numpy as np
date = np.array('2004-02-16', dtype=np.datetime64)
date
```

```python
date + np.arange(12)
```

1. _Because of the uniform type in NumPy datetime64 arrays, this type of operation can be accomplished much more quickly than if we were working directly with Python's datetime objects, especially as arrays get large_

2. _Pandas Timestamp object combines the ease-of-use of python datetime with the efficient storage and vectorized interface of numpy.datetime64_

3. _From a group of these Timestamp objects, Pandas can construct a DatetimeIndex that can be used to index data in a Series or DataFrame_

**DatetimeIndex Object**

* _A collection of pandas timestamp_

```python
# from Strings

pd.DatetimeIndex(['2004/2/16','2005/2/16','2006/2/16'])
# for storing multiple date time we use datetimeindex
```

```python
pd.DatetimeIndex(['2004/2/16','2005/2/16','2006/2/16'])[0]
# showing or storing the one datetime obj we use timestamps

```

```python
# using python datetime object
pd.DatetimeIndex([dt.datetime(2026, 1,1), dt.datetime(2026, 1,3),dt.datetime(2026, 1,2)])
```

```python
# using pd.Timestamp
dt_index = pd.DatetimeIndex([pd.Timestamp(2026, 1,1), pd.Timestamp(2026, 1,3),pd.Timestamp(2026, 1,2)])
```

```python
# using datetimeindex as series index
pd.Series([1,2,3], index=dt_index)
```
