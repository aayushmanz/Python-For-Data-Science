```python
import numpy as np 
import pandas as pd 
import seaborn as sns
```

```python
df = sns.load_dataset('tips')
df.head()
```

```python
df.groupby('sex')[['total_bill']].mean()
```

```python
# my way to find out the smoker people spend.

new = df[(df['smoker'] == 'Yes')]
new.groupby('sex')[['total_bill']].mean()
```

```python
# sir way :

df.groupby(['sex', 'smoker'])[['total_bill']].mean().unstack()
```

```python
# aggfunc() :

df.pivot_table(index='sex', columns='smoker', values='total_bill', aggfunc='sum')
```

```python
# all cols together :
df.pivot_table(index='sex', columns='smoker', values=['size', 'total_bill']) 
# we need to use values= in new version.
```

```python
# multidimentional :
df.pivot_table(index=['sex','smoker'],
               columns=['day','time'],
               values=['size','tip','total_bill'],
               aggfunc={'size':'mean', 'tip':'max', 'total_bill':'sum'})
```

```python
# margins :
df.pivot_table(index='sex', columns='smoker', values='total_bill', aggfunc='sum', margins=True)
```
